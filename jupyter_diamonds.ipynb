# I Python ska du nu:
a) Skapa en klass som heter BankAccount. Klassen skall ha attributet "account_holder"
som visar kontoinnehavarens namn samt attributet "balance" som visar
kontoinnehavarens balans. Klassen skall ha metoden "deposit()" för att kunna sätta in
pengar på kontot samt metoden "withdraw()" för att kunna ta ut pengar från kontot. Om
bankinnehavaren försöker ta ut mer pengar än vad som finns på kontot skall
meddelandet "Too low balance" printas ut.

b) Skapa en instans av klassen och testa så klassen funkar så som du förväntar dig. Du
kan till exempel prova printa ut attributen, sätta in pengar och ta ut pengar.


```python
class BankAccount():
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, cash):
        self.balance += cash
        return self.balance

    def withdraw(self, cash):
        if  self.balance - cash > 0:
            self.balance -= cash
            return self.balance
        else:
            return print("Too low balance")


Account = BankAccount("Tim", 1000)
print(Account.balance)

print(Account.deposit(100))
Account.withdraw(1200)
print(Account.withdraw(800))
```

# På intervjuer brukar ibland kodtest genomföras. Här är ett exempel på en fråga som hade
kunnat dyka upp på ett kodtest.
Skapa en Python funktion som beräknar antalet vokaler i en sträng. Vokalerna är:
A, E, I, O, U, Y, Å, Ä och Ö.
2/3
Exempelvis: vowel_checker(’hjärna’) ska returnera ”2” eftersom ordet ”hjärna” har de två
vokalerna ”ä” och ”a”. 

```python
def vowel_check(word):
    vowel_list = [ 'A', 'E', 'I', 'O', 'U', 'Y', 'Å', 'Ä', 'Ö']
    counter = 0

    for x in word.upper():
        if x in vowel_list:
            counter += 1

    return counter


print(vowel_check('AEIMMN'))
```

# På intervjuer brukar ibland kodtest genomföras. Här är ett exempel på en fråga som hade
kunnat dyka upp på ett kodtest.
Skapa en funktion som tar två listor som argument och returnerar de gemensamma
elementen i listorna i en ny lista.
Exempelvis:
common_elements([4, ’apple’, 10, ’hi’, 3], ([23 ’apple’, 5, 9, 3]) ska returnera svaret
[’apple’, 3].


```python
def combine_list(list1, list2):
    common = list(set(list1) & set(list2))
    return common

a = [1, 2, 3, 'hej', 5]
b = [1, 2, 6, 7, 'hej']

combine_list(a, b)
```

# Diagrams

```python
plt.scatter(data_diamonds['carat'], data_diamonds['price'])
plt.xlabel('carat')
plt.ylabel('price')
plt.title('Scatterplot: carat vs price')
plt.show()
```

The first thing i wanted to explore was if the carat (weight of the diamond) had a strong influence of the price. From the scatter plot we can see that the price generally increases with the carat going up.
Most diamonds in the dataset fall between the 1 - 2.2 carats but even within the range the price vary widely from 5000 to 18 000. 
This shows that while carat has a part in the price of the diamond other factors such as color, clarity and cut also plays a big part in determining the price.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

axes = axes.flatten()

data_diamonds.boxplot(column='price', by='cut', ax=axes[0])
axes[0].set_title('Price depending on cut')
axes[0].set_xlabel('Cut')
axes[0].set_ylabel('Price')

data_diamonds.boxplot(column='price', by='color', ax=axes[1])
axes[1].set_title('Price depending on color')
axes[1].set_xlabel('Color')
axes[1].set_ylabel('Price')

data_diamonds.boxplot(column='price', by='clarity', ax=axes[2])
axes[2].set_title('Price depending on clarity')
axes[2].set_xlabel('Clarity')
axes[2].set_ylabel('Price')

axes[3].axis('off')

plt.suptitle('Price depending on cut, color and clarity')
```

In these boxplots we analyze how cut, color and clarity affect the price of the diamond.

A clear observationen is that there is a large numbers of outliers across all categories especially the higher the grade is.
This suggests that the high quality can lead to higher prices the variation is also much greater likely due to other factors such as carat size.

For example diamonds with an Ideal cut which considered the best suprisingly shows a lower median price than the other cuts. This indicates that the cut alone doesn't drive the price up, perhaps the Ideal cut are more common or paired with lower grades in other qualities such as clarity and color.

Similar with diamonds the grade D which is the best has also one of the lowest median prices, yet shows a very wide spread in price and outliers.
While a lower tier grade J shows a more stable price range with fewer outliers suggesting more consistent pricing among the lower quality stones.

The same trend is visible in clarity, lower grade diamonds like I1 have a tighter price range while high grade such as IF have a larger price spread and more outliers.

Overall this suggests that higher quality diamonds have more volative pricing, likely because small differences in quality or carat can lead to price increases. Lower tier diamonds seems to be priced more consistently.

```python
data_diamonds['price_per_carat'] = data_diamonds['price'] / data_diamonds['carat']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

axes = axes.flatten()

data_diamonds.boxplot(column='price_per_carat', by='cut', ax=axes[0])
axes[0].set_title('Price per carat depending on cut')
axes[0].set_xlabel('Cut')
axes[0].set_ylabel('Price per carat')

data_diamonds.boxplot(column='price_per_carat', by='color', ax=axes[1])
axes[1].set_title('Price per carat depending on color')
axes[1].set_xlabel('Color')
axes[1].set_ylabel('Price per carat')

data_diamonds.boxplot(column='price_per_carat', by='clarity', ax=axes[2])
axes[2].set_title('Price per carat depending on clarity')
axes[2].set_xlabel('Clarity')
axes[2].set_ylabel('Price per carat')

axes[3].axis('off')

plt.suptitle('Price per carat depending on cut, color and clarity')
```

In these boxplots we explore how the price per carat varies depending on diamonds cut, color and clarity.

Compared to the earlier plots based on total price we now observe that price per carat is more stable accross all quality levels. The boxes IQR are narrower and there are fewer extreme outlier, this can indicate that using price per carat reduces variability and provides a more consistent basis for comparison.

When analyzing cut we notice that median price per carat remains fairly similar across all categories. However the ideal cut which is the highest grade shows more upward outliers. This suggest that while the average is stable some ideal diamonds are priced signifcantly higher per carat likely due to combinations with other high quality features.

A similar trend appears with color, diamonds with the best color grade "D" have slightly higher spread and more outliers while lower grade such as "J" show mroe consistent pricing with fewer outliers. This supports my idea that lower quality diamonds are price more stable while high quality ones vary more.

Regarding clarity we observe the most noticeable change compared to previous plots.
The IF category still has a wide spread and many outliers but most other clarity grades like VS1, VS2 and S1 now show less spread and fewer outliers indicating greater price consistency per carat.

My conclusion is that switching from total price to price per carat shows that this metric provides a clearer and more stable way to evalute diamond pricing. It smooths out variability related to diamond size and allows for more accurate comparison between quality grades. This makes price per carat a great tool for further analysis and for identifying diamonds that offer high value for money.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

median_price_per_cut = data_diamonds.groupby('cut')['price_per_carat'].median().sort_values()
median_price_per_color = data_diamonds.groupby('color')['price_per_carat'].median().sort_values()
median_price_per_clarity = data_diamonds.groupby('clarity')['price_per_carat'].median().sort_values()

axes = axes.flatten()

median_price_per_color.plot(kind='bar', color='skyblue', ax=axes[0])
axes[0].set_title('Price per carat depending on color')
axes[0].set_xlabel('color')
axes[0].set_ylabel('Median price per carat by color')

median_price_per_clarity.plot(kind='bar', color='skyblue', ax=axes[1])
axes[1].set_title('Price per carat depending on clarity')
axes[1].set_xlabel('Clarity')
axes[1].set_ylabel('Median Price per carat by clarity')

median_price_per_color.plot(kind='bar', color='skyblue', ax=axes[2])
axes[2].set_title('Price per carat depending on cut')
axes[2].set_xlabel('Color')
axes[2].set_ylabel('Median Price per carat by color')

axes[3].axis('off')

plt.suptitle('Median Price per carat depending on cut, color and clarity')
```

In these bar plots we show the median price per carat grouped by cut, color and clarity.

The goal of these graphs is to provide a clearer overview of how the median value changes depending on each quality grade. Unlike boxplots, which also show distribution and outliers, these bar plots focus specifally on median price helping us quickly identify which quality levels are generally associated with higher or lower price per carat values.

```python
#Higher number equals higher quality based on the type of category

cut_scores = {'Ideal': 5, 'Premium': 4, 'Very Good': 3, 'Good': 2, 'Fair': 1}
color_scores = {'D':7 ,'E':6 ,'F':5 ,'G': 4,'H': 3, 'I': 2,'J': 1}
clarity_scores = {'IF': 8,'VVS1': 7,'VVS2': 6,'VS1': 5,'VS2': 4,'SI1': 3,'SI2': 2,'I1': 1}

#Making new columns with scores based on the quality

data_diamonds['cut_scores'] = data_diamonds['cut'].map(cut_scores)
data_diamonds['color_scores'] = data_diamonds['color'].map(color_scores)
data_diamonds['clarity_scores'] = data_diamonds['clarity'].map(clarity_scores)

#Adding the scores together per diamond

data_diamonds['quality_total_score'] = data_diamonds['cut_scores'] + data_diamonds['color_scores'] + data_diamonds['clarity_scores']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_diamonds['quality_total_score'], data_diamonds['price_per_carat'])
plt.title('Price per carat based on quality score')
plt.xlabel('Quality score (color, clarity, cut)')
plt.ylabel('price per carat')
plt.grid(True)
plt.show()
```

This scatter plot will show based on the quality score the diamonts get the price will vary.

From my observation we can see that the higher the quality score that diamonds get the price per carat will increase, however as we can see from 16 to 19 we can observe that the price doesn't steadily increase a hundred procent since the dots above 12500 price per cart are mostily outliers.

My conclusion is that based on the graph we can asume that around the 15 to 19 range there are a lot of diamond with the same quality however some are outliers indacting that length, width or depth may play in part in those diamonds being more expensive.

```python
data_diamonds['value_score'] = data_diamonds['quality_total_score'] / data_diamonds['price_per_carat']

#Grouping all diamonds that have the same qualities into groupes and then taking the median value
median_value = data_diamonds.groupby(['cut', 'color', 'clarity', 'quality_total_score'])['value_score'].median().reset_index()

top_value = median_value.sort_values(by='value_score', ascending=False).head(10)
```
